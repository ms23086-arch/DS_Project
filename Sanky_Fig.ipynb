```python
import pandas as pd
import plotly.graph_objects as go

file_path = "/content/drive/MyDrive/Colab Notebooks/TEST.csv"
df = pd.read_csv(file_path)


df = df.dropna(subset=["NAME", "PHD", "POSTDOC", "INSTITUITE"])


labels = list(pd.unique(df[["NAME", "PHD", "POSTDOC", "INSTITUITE"]].values.ravel()))
label_to_index = {label: i for i, label in enumerate(labels)}


 #(1) NAME → PHD
links1 = pd.DataFrame({
    "source": df["NAME"].map(label_to_index),
    "target": df["PHD"].map(label_to_index),
    "value": 1
})

# (2) PHD → POSTDOC
links2 = pd.DataFrame({
    "source": df["PHD"].map(label_to_index),
    "target": df["POSTDOC"].map(label_to_index),
    "value": 1
})

# (3) POSTDOC → INSTITUTE
links3 = pd.DataFrame({
    "source": df["POSTDOC"].map(label_to_index),
    "target": df["INSTITUITE"].map(label_to_index),
    "value": 1
})


links = pd.concat([links1, links2, links3])


links = links.groupby(["source", "target"], as_index=False).agg({"value": "sum"})


fig = go.Figure(data=[go.Sankey(
    node=dict(
        label=labels,
        pad=20,
        thickness=20,
        line=dict(color="black", width=0.5)
    ),
    link=dict(
        source=links["source"],
        target=links["target"],
        value=links["value"]
    )
)])

fig.update_layout(
    title_text="NAME → PHD → POSTDOC → INSTITUITE Flow",
    font_size=10
)

fig.show()
```
